```python
# define list of different colours 
colours=["red", "blue", "yellow", "green", "orange", "white", "brown", "black", "gold", "cyan", "magenta", "silver", "pink", "crimson", "coral", "maroon", "olive", "plum", "sepia", "turquoise", "violet", "purple"]
print(len(colours))
```

```python
# define sequence in which you select route
route=['M1', 'M2', 'M4', 'M5', 'M6', 'M8', 'M10', '12', 'M13', '16', 'M17', '18', '21', '27', '37', '50', '60', '61', '62', '63', '67', '68']
```

```python
# Empty dictionary to store route and its colour
Assignment={'M1': '', 'M2': '', 'M4': '', 'M5': '', 'M6': '', 'M8': '', 'M10': '', '12': '', 'M13': '', '16': '', 'M17': '', '18': '', '21': '', '27': '', '37': '', '50': '', '60': '', '61': '', '62': '', '63': '', '67': '', '68': ''}
print(Assignment)
```

```python
N=len(df)
print(N)
no=len(route)
last=0
Assignment['M1']=colours[last]
for i in range(no):
  # list of feasible routes
  feasible_colours=colours.copy()
  # write code here for checking for intersection
  for j in range(i):
    flag=0
    if(df.iat[i,j+1]!=1):
      col=Assignment[route[j]]
      c=0
      for k in range(i):
        if col==Assignment[route[k]]:
          if df.iat[i,k+1]==1:
            c=c+1
      if c==0:
        flag=1
        Assignment[route[i]]=col
        break
    if flag==0 and j==i-1:
      last=last+1
      Assignment[route[i]]=colours[last]

  # write code here for deleting some colours from feasible_colours
feasible_colours=feasible_colours[0:last+1]  

  # write code here for storing route and its colour in Assignment dictionary
print(Assignment)  
print(feasible_colours)
print("Number of colours needed are:", len(feasible_colours))
```
